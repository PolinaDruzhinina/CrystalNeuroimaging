```python
PATH_TO_MRI = '/data'
i = 0 

for sub in os.listdir(PATH_TO_MRI):
    if len(sub) == 6:
        full_path = '{}/{}/unprocessed/3T/T1w_MPR1/{}_3T_T1w_MPR1.nii.gz'.format(PATH_TO_MRI, sub, sub)
        if os.path.exists(full_path):
            out_path = 'hcp_t1_fsl/{}_3T_T1w_MPR1_bet.nii'.format(sub)
            skullstrip = fsl.BET(in_file=full_path, out_file=out_path,
                mask =True)
            res = skullstrip.run()
            i=+1
            print(i)
```
